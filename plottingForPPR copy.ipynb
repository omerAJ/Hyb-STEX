```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from lib.metrics import test_metrics


def test(gumbell_model, best_model, dataloader, scaler, graph, args):
    gumbell_model.eval()
    best_model.eval()
    
    y_pred_g = []
    y_pred_best = []
    y_true = []
    
    with torch.no_grad():
        prev_len = 0
        for phase in ["test"]:
            for batch_idx, (data, target, evs, _) in enumerate(dataloader[phase]):
                
                def pred_model_g(data):
                    repr1, _ = gumbell_model(data, graph)                
                    pred_output = gumbell_model.predict(repr1, repr1, phase="pred")
                    return pred_output
                
                def pred_best_model(data):
                    repr1, _ = best_model(data, graph)                
                    pred_output = best_model.predict(repr1, repr1, phase="bias")
                    return pred_output

                
                pred_output_g = pred_model_g(data)
                pred_output_best = pred_best_model(data)
                y_true.append(target)
                y_pred_g.append(pred_output_g)
                y_pred_best.append(pred_output_best)
                
            print(f"Finished {phase} phase, with samples: {len(torch.cat(y_true, dim=0))-prev_len}")
            prev_len = len(torch.cat(y_true, dim=0))

    y_true = scaler.inverse_transform(torch.cat(y_true, dim=0)).cpu().numpy()
    y_pred_g = scaler.inverse_transform(torch.cat(y_pred_g, dim=0)).cpu().numpy()
    y_pred_best = scaler.inverse_transform(torch.cat(y_pred_best, dim=0)).cpu().numpy()
    test_results = []
    # inflow
    mae, mape = test_metrics(y_pred_best[..., 0], y_true[..., 0])
    print("\n\nbest\ntest, MAE: {:.2f}, test MAPE: {:.4f}%".format(mae, mape*100))
    test_results.append([mae, mape])
    mae, mape = test_metrics(y_pred_best[..., 1], y_true[..., 1])
    print("OUTFLOW, MAE: {:.2f}, MAPE: {:.4f}%".format(mae, mape*100))
    
    mae, mape = test_metrics(y_pred_g[..., 0], y_true[..., 0])
    print("\n\ngumbell\ntest, MAE: {:.2f}, test MAPE: {:.4f}%".format(mae, mape*100))
    test_results.append([mae, mape])
    mae, mape = test_metrics(y_pred_g[..., 1], y_true[..., 1])
    print("OUTFLOW, MAE: {:.2f}, MAPE: {:.4f}%".format(mae, mape*100))
    # test_results.append([mae, mape]) 

    return np.stack(test_results, axis=0), y_true, y_pred_g, y_pred_best
```

```python
_, y_true, y_pred_g, y_pred_best = test(gumbell_model, best_model, dataloader, dataloader['scaler'], graph, args)
```
